```python
# Ex 12 from here
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("spectrum_no_header.csv", sep=r"\s+", header=None, names=["Wellenzahl", "Intensitaet"])
print(df)

df["Intensitaet"] = df["Intensitaet"] * 20


df2 = pd.read_csv("spectrum_no_header_2.csv", sep=r"\s+", header=None, names=["Wellenzahl", "Intensitaet"])
print(df2)


plt.plot(df["Wellenzahl"], df["Intensitaet"], color="red", label="Spektrum 1 x20")
plt.plot(df2["Wellenzahl"], df2["Intensitaet"], color="blue", label="Spektrum 2")
plt.xlabel(r"Wellenzahl / cm$^{-1}$")
plt.ylabel("Intensitaet / a.u.")
plt.title(r"Schwingungsspektrum von C$_{120}^-$")
plt.xlim((0,2000))
plt.legend(loc="upper left")
plt.show
```

```python
# Ex 14 from here on
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

df = pd.read_csv("spectrum_no_header.csv", sep=r"\s+", header=None, names=["Wellenzahl", "Intensitaet"])
df["Intensitaet"] = df["Intensitaet"] * 20


df2 = pd.read_csv("spectrum_no_header_2.csv", sep=r"\s+", header=None, names=["Wellenzahl", "Intensitaet"])


df3 = pd.read_csv("spectrum3.csv", sep=r"\s+", header=None, names=["Wellenzahl", "Intensitaet"])


df4 = pd.read_csv("spectrum4.csv", sep=r"\s+", header=None, names=["Wellenzahl", "Intensitaet"])


fig, axs = plt.subplots(nrows=4, ncols=1, sharex=True)
axs[0].plot(df["Wellenzahl"], df["Intensitaet"], color="red", label="Spektrum 1 x20")
axs[1].plot(df2["Wellenzahl"], df2["Intensitaet"], color="blue", label="Spektrum 2")
axs[2].plot(df3["Wellenzahl"], df3["Intensitaet"], color="orange", label="Spektrum 3")
axs[3].plot(df4["Wellenzahl"], df4["Intensitaet"], color="green", label="Spektrum 4")
axs[0].set_ylabel("Intensitaet / a.u.")
axs[3].set_xlabel(r"Wellenzahl / cm$^{-1}$")
axs[3].set_xlim((0, 2000))
axs[0].legend()
axs[1].legend()
axs[2].legend()
axs[3].legend()

for ax in axs:
    ax.xaxis.set_minor_locator(MultipleLocator(50))
    ax.yaxis.set_minor_locator(MultipleLocator(1e-18))


```

```python
# Ex 15 and 16 from here 
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

df = pd.read_csv("spectrum.csv", sep=r"\s+", names=["Wellenzahl", "Intensitaet"], skiprows=1)

df["Intensitaet"] *= 20

plt.plot(df["Wellenzahl"], df["Intensitaet"], color="green", label="Spektrum 1")
plt.xlabel(r"Wellenzahl / cm$^{-1}$")
plt.ylabel("Intensitaet / a.u.")
plt.title(r"Schwingungsspektrum von C$_{120}^-$")
plt.xlim((0,2000))
plt.legend(loc="upper left")

ax = plt.gca()

# X-axis
ax.xaxis.set_major_locator(MultipleLocator(500))
ax.xaxis.set_minor_locator(MultipleLocator(100))

# y-axis
ax.yaxis.set_major_locator(MultipleLocator(1e-17))
ax.yaxis.set_minor_locator(MultipleLocator(2e-18))

# fonts
plt.rcParams.update(
    {
        "font.size": 10,
        "lines.linewidth": 1.5, 
        "xtick.major.size": 3.5,
        "xtick.minor.size": 2.0,
        "ytick.major.size": 3.5,
        "ytick.minor.size": 2.0,
        "axes.linewidth": 0.8,
    }
)

plt.show
```
