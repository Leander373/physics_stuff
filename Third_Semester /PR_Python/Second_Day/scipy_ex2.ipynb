```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load csv
baro = pd.read_csv("Barometer.csv")
#print(baro)

# Compute gradient
height = np.array(baro["relativeAltitude"])

velocity = np.gradient(height, np.array(baro["seconds_elapsed"]))
#print(velocity)

min_velocity = np.min(velocity)
max_velocity = np.max(velocity)
print(f"Max vertical velocity is {max_velocity}")
print(f"Min vertical velocity is {min_velocity}")


# First plot 
plt.plot(baro["seconds_elapsed"], baro["relativeAltitude"], color="green", label="Height")
# Second plot 
velocity *= 100
plt.plot(baro["seconds_elapsed"], velocity, color="orange", label="Vertical velocity x100")
plt.xlabel("Seconds elapsed")
plt.ylabel("relative Altitude / m/s")
plt.legend(loc="lower left")
```
