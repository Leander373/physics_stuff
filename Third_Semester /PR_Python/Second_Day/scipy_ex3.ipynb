```python
import numpy as np
from scipy.optimize import approx_fprime
import matplotlib.pyplot as plt


def derivative(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

def main():
    f = lambda x: np.sin(x)
    h = 1e-6
    n_points = 1000
    max_diff = np.zeros(n_points)

    # Array for h steps
    steps = np.zeros(n_points)

    # Loop for several h 
    for i in range(n_points):
        x = 0
        max_err = 0
        steps[i] = h

        # For interval from 0 to 2 pi
        while x < (2 * np.pi):
            current = abs(derivative(f, x, h) - np.cos(x))
            if current > max_err:
                max_diff[i] = current
                max_err = current
            x += (1/1000)

        h += 1e-6
    
    # Array for difference 
    proximity = np.array(max_diff)
    h_steps = np.array(steps)


    # Plot 
    plt.scatter(h_steps, proximity, color="pink", marker="o", s=15)
    plt.xlabel("Values for h")
    plt.ylabel("Max Difference")
    plt.title("Maximum difference of central differential and derivative")
    plt.grid(True, which="both", linestyle=":")
    
    plt.show()

    





main()
```
